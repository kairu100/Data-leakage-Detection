**Outlier Detection using Isolation Forest Algorithm on Credit Card Dataset**

Using the Isolation Forest algorithm for identifying outliers in the credit
card dataset. Isolation Forest is an unsupervised learning algorithm designed for
efficient anomaly detection.


```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

def detect_anomalies(data):
    # Features and target variable
    X = data.drop('card', axis=1)  # Features (excluding the target column 'card')

    # Ensure that X has valid feature names
    X.columns = [f'feature_{i}' for i in range(X.shape[1])]

    # Initializing the IsolationForest model
    model = IsolationForest(random_state=42, contamination=0.5)  # You can adjust the contamination parameter

    # Fit the IsolationForest model on the data
    model.fit(X)

    # Predictions (-1 for anomalies, 1 for inliers)
    predictions = model.predict(X)

    # Display the predicted instances
    print("Predicted Anomalies:")
    print(predictions)

    # Get the indices of the anomalies
    anomaly_indices = data[predictions == -1].index

    # Extract anomaly instances
    anomaly_instances = data.loc[anomaly_indices]

    # Remove anomaly instances from the data
    cleaned_data = data.drop(anomaly_indices)

    # Calculate the number of anomalies identified
    num_anomalies = (predictions == -1).sum()
    total_samples = len(predictions)
    percentage_anomalies = (num_anomalies / total_samples) * 100

    print(f"Number of anomalies: {num_anomalies} out of {total_samples} samples")
    print(f"Percentage of anomalies: {percentage_anomalies:.2f}%")

    return anomaly_instances, cleaned_data, percentage_anomalies

# Assuming the data provided is stored in a DataFrame named 'data_iso'
data_iso = pd.DataFrame({
    'card': [True, True, True, True, True],
    'reports': [0, 0, 0, 0, 0],
    'age': [37.66667, 33.25, 33.66667, 30.5, 32.16667],
    'income': [4.52, 2.42, 4.5, 2.54, 9.7867],
    'share': [0.03327, 0.005217, 0.004156, 0.065214, 0.067051],
    'expenditure': [124.9833, 9.854167, 15.0, 137.8692, 546.5033],
    'owner': [True, False, True, False, True],
    'selfemp': [False, False, False, False, False],
    'dependents': [3, 3, 4, 0, 2],
    'months': [54, 34, 58, 25, 64],
    'majorcards': [1, 1, 1, 1, 1],
    'active': [12, 13, 5, 7, 5]
})

# Detect anomalies and display additional information
anomaly_instances_iso, cleaned_data_iso, percentage_anomalies_iso = detect_anomalies(data_iso)

print("\nAnomaly instances:")
print(anomaly_instances_iso)

print("\nData without anomalies:")
print(cleaned_data_iso)

```

**1 (Positive Label):** Represents inliers or normal instances. Inliers are expected to have a longer average path length in the isolation trees, meaning they are harder to separate from the majority of the data.

**-1 (Negative Label):** Represents outliers or anomalous instances. Outliers are expected to have a shorter average path length in the isolation trees, indicating that they are easier to isolate.
The decision to use 1 and -1 is related to the scoring mechanism used by the Isolation Forest. **Normal Data Points**: Anomaly scores close to 1.
**Anomalous Data Points:** Anomaly scores close to -1.
During the training phase, the algorithm constructs a forest of isolation trees. The average path length for each instance is calculated, and during the prediction phase, instances with shorter average path lengths are considered more likely to be anomalies.

By assigning labels of 1 and -1, the algorithm can use a simple rule: if the predicted label is -1, the instance is considered an anomaly; if the label is 1, the instance is considered normal. This binary labeling simplifies the interpretation of the predictions.

While the choice of 1 and -1 might seem arbitrary, it aligns with the goal of the algorithm to efficiently identify anomalies based on their isolation characteristics in the feature space.

**The Isolation Forest algorithm** flagged the instance at **index 4** as an anomaly in the credit card dataset due to its significantly **higher expenditure** **(546.5033)** and **share value (0.067051)**. These features, along with a longer **card usage duration ('months' = 64)** and **card ownership ('card' = True)**, contributed to the instance being isolated as an anomaly. The algorithm identifies anomalies based on the distinctiveness of multiple features, and in this case, the combination of higher expenditure and share values played a crucial role in flagging the instance as anomalous.

The threshold for identifying anomalies in the dataset is controlled by the contamination parameter, which you set to 0.2 in your code. This parameter represents the proportion of outliers or anomalies expected in the dataset. You have the flexibility to adjust this parameter based on your specific use case and the level of sensitivity you desire in detecting anomalies. A higher contamination value makes the model more tolerant to outliers, while a lower value increases the model's sensitivity, potentially identifying more instances as anomalies.

**Identifying Potential Leaky Predictors in Time-Dependent Data**

Leaky
predictors are features that may inadvertently contain information from the future,
risking data leakage during model training.


```python
# Features and target variable
X = data.drop('card', axis=1)  # Features (excluding the target column 'card')
y = data['card']  # Target variable 'card'

# Assuming a single time shift for comparison
time_shift = 1

leaky_predictors = []

for column in X.columns:
    shifted_column = X[column].shift(-time_shift).dropna()  # Exclude rows with NaN
    match_indices = (shifted_column == y.iloc[:-time_shift])

    if match_indices.any():  # Check if any match is found
        leaky_predictors.append(column)

if leaky_predictors:
    print("Potential leaky predictors found:", leaky_predictors)
else:
    print("No potential leaky predictors identified.")
```

The code is identifying potential leaky predictors, specifically 'owner' and 'majorcards,' in the dataset. Leaky predictors inadvertently incorporate information from the future into the model, leading to overly optimistic performance evaluations. The output indicates that the columns 'owner' and 'majorcards' exhibit the same values as their respective columns shifted by one time step, suggesting a potential issue. Including such features may result in overfitting, where the model learns patterns based on future data, hindering its ability to generalize effectively to unseen instances.

**Debugging Predicted vs. Actual Values for Leaky Predictor Identification**

This aids in the identification and understanding of potential leaks,
helping to prevent unintentional exposure of future information to the model..

```python
# Actual target values (ground truth)
actual_targets = y.shift(-time_shift).dropna()

# Initialize predicted_leaky with zeros
predicted_leaky = pd.Series(0, index=actual_targets.index)

# Set values to 1 where leaky predictors are identified
for leaky_predictor in leaky_predictors:
    shifted_column = X[leaky_predictor].shift(-time_shift).dropna()
    match_indices = (shifted_column == y.shift(-time_shift).dropna())
    predicted_leaky[match_indices.index] = 1

# Calculate accuracy
accuracy = (predicted_leaky == actual_targets).mean() * 100

print("Predicted Leaky Values:\n", predicted_leaky.values)
print("Actual Targets Values:\n", actual_targets.values)
```

The output indicates the identification of potential leaks in the model, with the "Predicted Leaky Values" being [1 1 1 1] and the "Actual Targets Values" as [True True True True]. This implies that the model has identified leaks for each corresponding row where the actual target values are True. The "1" in the "Predicted Leaky Values" suggests a match between shifted predictor values and shifted target values, signifying a potential pattern learned by the model that could be indicative of information leakage
